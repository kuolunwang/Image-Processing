```python
# library
import numpy as np
import cv2
from matplotlib import pyplot as plt
from google.colab.patches import cv2_imshow
```

```python
# load picture from google drive
from google.colab import drive
drive.mount('/content/drive')
```

```python
# read image and transfer to gray
img = cv2.imread('/content/drive/My Drive/Bird feeding 3 low contrast.tif',0)
```

```python
# s = T(r)
s = ((np.arctan((img - 128.0) / 32)+1.325817) * 96.2343).astype(np.uint8)
```

```python
# show processed image
cv2_imshow(s)
```

```python
plt.hist(img.ravel(),256,[0,256])
plt.title("Histogram of origin image")
```

```python
plt.hist(s.ravel(),256,[0,256])
plt.title("Histogram of processed image")
```

```python
# show transfer function

x = np.arange(256)
y = ((np.arctan((x - 128.0) / 32)+1.325817) * 96.2343).astype(np.uint8)
plt.plot(x,y)
plt.axis([0,255,0,255])
plt.title("transfer function")
plt.xlabel("Input gray level, r")
plt.ylabel("Output gray level, s")
```

```python
import pandas as pd

# combine input and output

dic = dict(zip(x,y))
index = ["output"]
```

```python
table = pd.DataFrame(dic,index)
table
```
